#### See the artifact 'emp-single-end-sequences.qza'

#### See the 'demux.qza' from 'Demultiplexing sequences' step

```python
qiime tools extract \
  --input-path demux.qza \
  --output-path demux
```

##### See, the seqs for each sample has been sperated.

#### See the 'table.qza' from 'Sequence quality control and feature table construction' step

```python
qiime tools extract \
  --input-path table.qza \
  --output-path table-dada2

# convert 'biom' to 'tsv'
biom convert -i table-dada2/6e85dda1-969f-4f65-8a8f-aa23cf7ad043/data/feature-table.biom -o table-dada2/6e85dda1-969f-4f65-8a8f-aa23cf7ad043/data/feature-table.tsv --to-tsv
```

##### This is the FeatureTable[Frequency], also named OTU table.

#### See the reconstructed trees from 'Generate a tree for phylogenetic diversity analyses' step

```python
# Export unrooted tree
qiime tools export \
  --input-path unrooted-tree.qza \
  --output-path unrooted-tree

# Export rooted tree
qiime tools export \
  --input-path rooted-tree.qza \
  --output-path rooted-tree
```

#### See the 'evenness_vector.qza' (alpha diversity) from 'Alpha and beta diversity analysis' step

```python
qiime tools export \
  --input-path evenness_vector.qza \
  --output-path evenness_vector
```

#### See the 'bray_curtis_distance_matrix.qza' (beta diversity) from 'Alpha and beta diversity analysis' step

```python
qiime tools export \
  --input-path bray_curtis_distance_matrix.qza \
  --output-path bray_curtis_distance_matrix
```
