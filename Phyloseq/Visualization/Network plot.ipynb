#### Prepare

```r
# Import packages
library("phyloseq")
library("ggplot2")

theme_set(theme_bw())
```

```r
# Import data
data("enterotype")
pseq <- subset_samples(enterotype, !is.na(Enterotype))

head(otu_table(pseq))
head(sample_data(pseq))
head(tax_table(pseq))
```

#### Plot

```r
# Basic plot
plot_net(pseq, distance = "bray", maxdist = 0.1, point_label = "Sample_ID")

# Define 'color' and 'shape'
plot_net(pseq, distance = "bray", maxdist = 0.1, color = "SeqTech", shape = "Enterotype")

# Change layout
plot_net(pseq, distance = "bray", maxdist = 0.1, color = "SeqTech", shape = "Enterotype", laymeth = "circle")
```
