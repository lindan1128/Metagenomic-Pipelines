#### Prepare

```r
# Import packages
library("phyloseq")
library("ggplot2")

theme_set(theme_bw())
```

```r
# Import data
data("GlobalPatterns")
gpt <- subset_taxa(GlobalPatterns, Kingdom=="Bacteria")
pseq <- prune_taxa(names(sort(taxa_sums(gpt),TRUE)[1:300]), gpt)

head(otu_table(pseq))
head(sample_data(pseq))
head(tax_table(pseq))
```

#### Plot

```r
# Basic plot
plot_heatmap(pseq)

# Group samples and define ordination, distance and color
plot_heatmap(pseq, method = "NMDS", distance = "bray", ample.label = "SampleType", taxa.label = "Family", low = "#FFFFCC", high = "#000033", na.value = "white")
```
