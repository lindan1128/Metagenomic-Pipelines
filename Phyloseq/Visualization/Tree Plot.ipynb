#### Prepare

```r
# Import packages
library("phyloseq")
library("ggplot2")

theme_set(theme_bw())
```

```r
# Import data
data("GlobalPatterns")
pseq <- prune_taxa(taxa_names(GlobalPatterns)[1:50], GlobalPatterns)

phy_tree(pseq)
```

#### Plot

```r
# Basic plot
plot_tree(pseq)

# Choose taxa
plot_tree(pseq, color = "Genus", label.tips = "Genus")

# Define 'abundance' as 'size'
plot_tree(pseq, size = "abundance", color = "SampleType", base.spacing = 0.03, min.abundance = 3)

# Define 'color' 'size' and 'shape'
plot_tree(pseq, nodelabf = nodeplotboot(80,0,3), size = "abundance", color = "SampleType", shape = "Class", label.tips = "taxa_names", ladderize = "left")

# Radial plot
plot_tree(pseq, nodelabf = nodeplotboot(60,60,3), color = "SampleType", shape = "Class", ladderize = "left") + coord_polar(theta = "y")
```
