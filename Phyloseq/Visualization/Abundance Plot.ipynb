#### Prepare

```r
# Import packages
library("phyloseq")
library("ggplot2")

theme_set(theme_bw())
```

```r
# Import data
data("GlobalPatterns")
pseq <- subset_taxa(GlobalPatterns, Phylum == "Chlamydiae")

head(otu_table(pseq))
head(sample_data(pseq))
head(tax_table(pseq))
```

#### Plot

```r
# Basic plot
plot_bar(pseq)

# Fill the color
plot_bar(pseq, fill = "Genus")

# Group samples
plot_bar(pseq, x = "SampleType", fill = "Genus")

# Use facets
plot_bar(pseq, x = "Family", fill = "Genus", facet_grid = ~SampleType)
```
