#### Prepare

```r
# Import packages
library("phyloseq")
library("ggplot2")

theme_set(theme_bw())
```

```r
# Import data
data("GlobalPatterns")
pseq <- prune_species(speciesSums(GlobalPatterns) > 0, GlobalPatterns)

head(otu_table(pseq))
head(sample_data(pseq))
head(tax_table(pseq))
```

#### Plot

```r
plot_richness(pseq)

plot_richness(pseq, measures = c("Chao1", "Shannon"))

plot_richness(pseq, x = "SampleType", measures = c("Chao1", "Shannon"))

plot_richness(pseq, x = "SampleType", color = "SampleType", measures = c("Chao1", "Shannon")) +
geom_point(size=5, alpha=0.7)
```
