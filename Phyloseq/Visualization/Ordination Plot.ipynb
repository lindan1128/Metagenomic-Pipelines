#### Prepare

```r
# Import packages
library("phyloseq")
library("ggplot2")
library("plyr")

theme_set(theme_bw())
```

```r
# Import data
data("GlobalPatterns")

wh0 = genefilter_sample(GlobalPatterns, filterfun_sample(function(x) x > 5), A = 0.5*nsamples(GlobalPatterns))
pseq = prune_taxa(wh0, GlobalPatterns)

head(otu_table(pseq))
head(sample_data(pseq))
head(tax_table(pseq))
```

#### For taxas

```r
ord <- ordinate(pseq, "NMDS", "bray")
p1 <- plot_ordination(pseq, ord, type = "taxa", color = "Phylum", title = "taxa")

print(p1)

p1 + facet_wrap(~Phylum)
```

#### For samples

```r
ord <- ordinate(pseq, "NMDS", "bray")
p2 <- plot_ordination(pseq, ord, type = "samples", color = "SampleType") 
p2 + geom_polygon(aes(fill = SampleType)) + geom_point(size = 5) + ggtitle("samples")

p2 + facet_wrap(~SampleType)
```

#### For biplot graphic

```r
ord <- ordinate(pseq, "NMDS", "bray")
p3 <- plot_ordination(pseq, ord, type = "biplot", color = "SampleType", shape = "Phylum", title="biplot")

print(p3)
```

#### For split graphic

```r
ord <- ordinate(pseq, "NMDS", "bray")
p4 <- plot_ordination(pseq, ord, type = "split", color = "Phylum", shape = "SampleType",title = "split") 

print(p4)
```
