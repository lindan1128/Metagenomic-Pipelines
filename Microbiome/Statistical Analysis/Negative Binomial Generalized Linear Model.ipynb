#### Prepare

```r
# Import packages
library("microbiome")
library("ggplot2")
library("dplyr")
library("IRanges")
library("lme4")
library("MASS")
```

```r
# Import data
data(peerj32) 
pseq <- peerj32$phyloseq 

head(otu_table(pseq))
head(sample_data(pseq))
head(tax_table(pseq))
```

#### Abundance boxplot

```r
p <- boxplot_abundance(pseq, x = "time", y = "Akkermansia", line = "subject") + 
     scale_y_log10()

print(p)
```

#### Negative binomial generalized linear model

```r
# Analyse specific taxa
tax <- "Akkermansia"

# Pick the signal (abundance) for this tax
sample_data(pseq)$signal <- get_sample(pseq, tax)

# Negative binomial test with group and gender included
res <- glm.nb(signal ~ group + sex, data = meta(pseq))

# Show the results
print(coef(summary(res)))
```
